* **H3 (Geografía):** Los puntajes de anomalía geográfica son críticos para detectar accesos no autorizados.

```python
import pandas as pd
import numpy as np
import requests # Para la API
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 1. Dirección del dataset original (GitHub Media)
url_original = 'https://media.githubusercontent.com/media/Pablo2710/ProyectoDSI/refs/heads/main/financial_fraud_detection_dataset.csv'

print("Cargando dataset original desde GitHub...")
df_original = pd.read_csv(url_original)
print(f"Dataset cargado exitosamente. Registros: {df_original.shape[0]}")

# 2. Configuración del Repositorio de Destino (DSII)
# Clonamos tu repo de la segunda parte para guardar los archivos allí
REPO_URL = "https://github.com/Pablo2710/ProyectoDSII.git"
REPO_NAME = "ProyectoDSII"

if not os.path.exists(REPO_NAME):
    !git clone {REPO_URL}
else:
    print(f"El repositorio {REPO_NAME} ya está clonado.")

%cd {REPO_NAME}
```

```python
# 1. Aplicamos el muestreo balanceado (2000 vs 2000)
fraude = df_original[df_original['is_fraud'] == 1]
no_fraude = df_original[df_original['is_fraud'] == 0]

df_balanced = pd.concat([
    fraude.sample(n=2000, random_state=42),
    no_fraude.sample(n=2000, random_state=42)
]).sample(frac=1, random_state=42).reset_index(drop=True)

# 2. Limpieza de Leakage (Requisito de la entrega anterior)
if 'fraud_type' in df_balanced.columns:
    df_balanced = df_balanced.drop(columns=['fraud_type'])

# 3. Guardar el archivo balanceado dentro de la carpeta del repo clonado
output_filename = "financial_fraud_balanced.csv"
df_balanced.to_csv(output_filename, index=False)
print(f"Archivo balanceado '{output_filename}' generado.")

# Nota: Para subirlo a GitHub desde Colab, necesitarás configurar tu User y Token.
# Por ahora, el archivo ya vive en la carpeta local de tu proyecto.
```
