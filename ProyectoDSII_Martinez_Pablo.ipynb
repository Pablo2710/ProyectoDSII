```python
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración visual
sns.set_theme(style="whitegrid")

# Requisito: Importar datos de una API (Tipo de cambio para enriquecer datos)
url_api = "https://api.exchangerate-api.com/v4/latest/USD"
try:
    response = requests.get(url_api)
    usd_to_ars = response.json()['rates']['ARS']
    print(f"✅ API Conectada. Tipo de cambio: 1 USD = {usd_to_ars} ARS")
except:
    usd_to_ars = 1000 # Valor de respaldo por si falla la API
    print("⚠️ Falló la API. Usando valor de respaldo.")
```

```python
# 1. Cargar datos originales desde tu repositorio DSI
url_github = 'https://media.githubusercontent.com/media/Pablo2710/ProyectoDSI/refs/heads/main/financial_fraud_detection_dataset.csv'
print("Cargando dataset original desde GitHub... (puede demorar unos segundos)")
df_original = pd.read_csv(url_github)

# 2. Separar clases para el Under-sampling
fraude = df_original[df_original['is_fraud'] == 1]
no_fraude = df_original[df_original['is_fraud'] == 0]

# 3. Extraer 2000 de cada uno y mezclar
df_balanced = pd.concat([
    fraude.sample(n=2000, random_state=42),
    no_fraude.sample(n=2000, random_state=42)
]).sample(frac=1, random_state=42).reset_index(drop=True)

# 4. Limpieza (Evitar Leakage) y Enriquecimiento (API)
if 'fraud_type' in df_balanced.columns:
    df_balanced = df_balanced.drop(columns=['fraud_type'])

df_balanced['amount_ARS'] = df_balanced['amount'] * usd_to_ars

# 5. Guardar archivo localmente (luego lo subís a mano a tu repo DSII)
df_balanced.to_csv('financial_fraud_balanced.csv', index=False)
print(f"✅ Dataset balanceado creado y guardado. Filas totales: {len(df_balanced)}")
```
